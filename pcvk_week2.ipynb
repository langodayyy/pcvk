````python
#membuat 5 box
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Fungsi untuk menghasilkan warna acak
def random_color():
    return np.random.randint(0, 256, size=3, dtype=np.uint8).tolist()

# Fungsi untuk menghasilkan ukuran dan posisi acak untuk kotak
def random_rect_params(image_size):
    x1 = np.random.randint(0, image_size[1] - 50)
    y1 = np.random.randint(0, image_size[0] - 50)
    x2 = np.random.randint(x1 + 20, image_size[1])
    y2 = np.random.randint(y1 + 20, image_size[0])
    return (x1, y1, x2, y2)

# Membuat gambar hitam dengan ukuran 512x512 piksel
image_size = (512, 512)
black_img = np.zeros((image_size[0], image_size[1], 3), dtype=np.uint8)

# Menggambar 5 kotak dengan ukuran dan warna acak
for _ in range(5):
    color = random_color()
    x1, y1, x2, y2 = random_rect_params(image_size)
    # Menggambar hanya garis kotak dengan ketebalan 2 piksel
    cv2.rectangle(black_img, (x1, y1), (x2, y2), color, 2)

# Menampilkan gambar
plt.imshow(black_img)
plt.title('Image with 5 Randomly Generated Boxes')
plt.show()
````
